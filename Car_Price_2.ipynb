```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df1 = pd.read_csv("CarPrice_Assignment.csv")
df1
```

```python
features = ["wheelbase","carlength","carwidth","carheight","curbweight","enginesize","boreratio","stroke","compressionratio","horsepower","peakrpm","citympg","highwaympg","price"]
data = df1[features].copy()
data
```

```python
X = data.copy()
X = data.drop('price', axis = 1) 
y = data.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
## Z-Score Normalisation - 

## for training set - 


X_train_scaled = (X_train - X_train.mean()) / X_train.std() 
y_train_scaled = (y_train -y_train.mean())/ y_train.std()

## for testing set - 

X_test_scaled = (X_test - X_test.mean())/ X_test.std()
y_test_scaled = (y_test - y_test.mean())/ y_test.std()




```

```python
## for training set  -
q=len(X_train)
r=len(X_train)

#adding a column of zeros
x=np.zeros((q,r))
x=X_train_scaled.to_numpy()

Y = y_train_scaled.to_numpy()
r
```

```python
num_epochs = 10 # number of iterations

z = np.zeros(r)

a = 0.1 # learning rate

cost_func =[]
```

```python
# cost function
def cost_function(z):
    
        n=np.dot(x,z)
        
        cost = 0.5*(1/q)*(np.sum(np.square(n-y))) ## cost function
        
        cost_func.append(cost)
```

```python
# gradient descent

for j in range(num_epochs):
    
    cost_function(z)
    
    n=np.dot(x,z)

    gra_des=(1/q)*np.dot((n-y),x)

    z=z-(a*gra_des)
```
