```python
train_d = pd.read_csv("CarPrice_Assignment.csv")
```

```python
features = ["wheelbase","carlength","carwidth","carheight","curbweight","enginesize","boreratio","stroke","compressionratio","horsepower","peakrpm","citympg","highwaympg","price"]
data = train_d[features].copy()
data
```

```python
x_train = data.copy()
x_train = data.drop('price', axis = 1) 
y_train = data.iloc[:,-1]
x_train
```

```python
## Z-Score Normalisation - 


x_train_scaled = (x_train - x_train.mean()) / x_train.std() 
y_train_scaled = (y_train -y_train.mean())/ y_train.std()

x_train_scaled

```

```python
## for training set  -
q=len(x_train)
r=len(x_train.loc[0])+1

#adding a column of zeros
x=np.zeros((q,r))
x[:,1:]=x_train_scaled.to_numpy()

y = y_train_scaled.to_numpy()
```

```python
num_epochs = 1000 # number of iterations

z = np.zeros(r)

a = 0.07 # learning rate

cost_func =[]
```

```python
# cost function
def cost_function(z):
    
        n=np.dot(x,z)
        
        cost = 0.5*(1/q)*(np.sum(np.square(n-y))) ## cost function
        
        cost_func.append(cost)
```

```python
# gradient descent

for j in range(num_epochs):
    
    cost_function(z)
    
    n=np.dot(x,z)

    gra_des=(1/q)*np.dot((n-y),x)

    z=z-(a*gra_des)
```

```python
plt.plot(list(range(num_epochs)),cost_func,color = "Blue")

plt.title("Cost Function after numerous epochs")

plt.xlabel("Epochs")

plt.ylabel("Cost")

plt.show()
```
