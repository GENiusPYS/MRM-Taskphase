```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

df=pd.read_csv("titanic.csv")
df
```

```python
df.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}},inplace= True)
X=df[["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
X["Age"].fillna(X["Age"].median(skipna=True), inplace=True)
X["Embarked"].fillna(X['Embarked'].value_counts().idxmax(), inplace=True)
Y=df["Survived"]
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))



class LogisticRegression:
    
     def __init__(self, lr=0.0005, it=2000):
        self.lr=lr
        self.it= it
        self.weights= None
        self.bias= None
    
     def optimize(self, x, y):
        sample, n = x.shape 
        self.weights = np.zeros(n) 
        self.bias = 0

        for i in range(self.it):  
            linear=np.dot(x,self.weights)+self.bias
            predictions=sigmoid(linear)
            
            dw = (1/sample) * np.dot(x.T, (predictions - y)) 
            db = (1/sample) * np.sum(predictions-y) 
            
            
            self.weights = self.weights - self.lr * dw 
            self.bias = self.bias - self.lr * db   
    
     def predict(self, x):
        linear=np.dot(x,self.weights)+self.bias
        predictions=sigmoid(linear)
        preds=[0 if y<=0.5 else 1 for y in predictions]
        return preds 
    
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
reg = LogisticRegression()
reg.optimize(X_train,Y_train)
predictions = reg.predict(X_test)
print(predictions)
```
